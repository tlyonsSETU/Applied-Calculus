## Introduction




* In this practical we will use Python to plot the __surface__ of a __solid of revolution__, which is generated when we __rotate__  some function $f(x)$ about the $x$-axis.
 
### Mathematics Concepts

 * The numerical value of the function $f$ at some point $x_0$ represents the height of the graph of the function above the $x$-axis. 
 
* When we rotate this function about the $x$-asis, the piece of the function located at $x_0$ traces out a disc of radius $f(x_0)$ about the $x-$axis. 

* Hence the area of this disc is $\pi f(x_0)^2$.

* When we multiply this disc area by the line-element $dx$ we obtain the volume of an intinitessimal volume segment, whose volumen is given by
$$
    dV = \pi f(x_0)^2dx.
$$

* When we integrate this volume element over an interval of the line $x\in(a,b)$, this gives the volume generated by the function segment between the points $x=a$ and $x=b$
$$
V=\int_{a}^{b}\pi f(x)^2dx \quad \text{(Volume of revolution)}
$$



### Example 1

* Generate a __surface plot__ of the volume generated by the function

$$
 f(x)=\sin(x)\quad  \text{ for } -\pi\leq x\leq \pi
$$

when it is rotated about the $x$-axis.


#### Part 1 - Setting up the coordinate system 

We firt set up our coordinate system, which will allow us to plot the surface: 

* The endpoints of the $x$-axis in this example are given by

```python
a1=-np.pi
b1=np.pi
x1=np.arange(a1,b1,0.01) # The x-axis for this example
```

* When we rotate the function about the $x$-axis we will rotate the function through a full circle, in other words through $2\pi$ radians.


* We call this coordinate $t$, and define it as follows

```python
t=np.arange(0,2*np.pi,0.01)
```

* We now create a __2-dimenstional grid__ from these two corrdinates using __meshgrid__ 

```python
X1,T=np.meshgrid(x1,t) 
```

*  We use **x and  t** to denote the coordinates on the 2-dimensoinal grid 

#### Part 2 - Input the generating function $f$

```python
f=np.sin(x1)
```

#### Part 3 - Plot the generating function in 3-dimensions

```python
fig=plt.figure(figsize=(12,12))
ax=fig.add_subplot(projection='3d') # Use a 3-d projection to plot the surface

ax.plot(x1,f,0,color='k') # plot the function in the (x,y)-plane i.e. z=0
ax.plot(x1,f*np.cos(0.8),f*np.sin(0.8),color='blue') # plot the function rotated by 0.8 rad about the x-axis

ax.plot([a1-1,b1+1],[0,0],[0,0],color='k',lw=1) # Plot the x-axis



ax.set_xlabel('x',size=12) # Label the x-axis
ax.set_ylabel('y',size=12) # Label the y-axis
ax.set_zlabel('z',size=12) # Label the z-axis

ax.view_init(22,7) # Change the viewing angle

plt.show()
```

* The three arguments of the __plot()__ funciton here are
    
  1. __x__ the coordinate long the $x$-axis
    
  2. __f__ the $y$-coordiante, i.e. $y=f(x)$
    
  3. The $z$-coordinate, i.e. $z=0$ (until we rotate the function about the $x$-axis)

#### Part 4 - Generating the surface plot

* We now create the $Y$ and $Z$ coordiantes for the surface plot:

* These are given by the following

```python
Y1=f*np.cos(T)
Z1=f*np.sin(T)
```

* The __surface plot__ itself is given by

```python
# Steps from Part 3
fig=plt.figure(figsize=(12,12))
ax=fig.add_subplot(projection='3d') # Use a 3-d projection to plot the surface
ax.plot(x1,f,zs=0,zdir='z',color='k') # plot the function in the (x,y)-plane i.e. z=0 (See Part 3)

# The Surface Plot
ax.plot_surface(X1,Y1,Z1,cmap=cm.Blues,linewidth=0.0,alpha=0.5) # Plot the surfcae of revolution generated by the curve
plt.show()
```

* The arguments of the __plot_surface()__ functions are as follows
    1. __X1__ the $x$-ccordinate of a point on the surface
    2. __Y1__ the $y$-coordinatee of a point on the surface
    3. __Z1__ the $z$-ccordinate of a point on the surface
    4. __cmap__ the __color map__ used to color the surface  
    (a full lise of color maps is avaailable at https://matplotlib.org/examples/color/colormaps_reference.html)
    5. __linewidth__ the width of the lines on the surface
    6. __alpha__ the transparency of the surface (smaller __alpha__ = more transparent)
    

#### Part 5 - Improving the plot

* We can imporve the plot by including labels for the coordinate axes using the functions
    * __ax.set_xlabel()__, __ax.set_ylabel()__, __ax.set_zlabel()__

* We can also use the function __ax.view_init()__ to change our viewing angle of the surface

```python
# Steps from Part 3
fig=plt.figure(figsize=(12,12))
ax=fig.add_subplot(projection='3d') # Use a 3-d projection to plot the surface
ax.plot(x1,f,zs=0,zdir='z',color='k') # plot the function in the (x,y)-plaane i.e. z=0 (See Part 3)

# Step from Part 4 -The Surface Plot
ax.plot_surface(X1,Y1,Z1,cmap=cm.Blues,linewidth=0.0,alpha=0.5) # Plot the surfcae of revolution generated by the curve


# Part 5 - Labels and Viewing angle
ax.set_xlabel('$x$') # Set the x-label 
ax.set_ylabel('$y$') # Set the y-label
ax.set_zlabel('$z$') # Set the z-label


ax.view_init(45,35) # Rotate your view of the solid

plt.show()
```

## Example 2

Plot the surface generated by the function
$$
g(x)=x^2 \quad \text{for }-5\leq x\leq 5
$$
as a __wireframe__.

#### Part 1 - Setting up the coordinate system

```python
a2=-5
b2=5
x2=np.arange(a2,b2,0.01)

t=np.arange(0,2*np.pi,0.01)

X2,T=np.meshgrid(x2,t)  
```

#### Part 2 - Create the generating function

```python
g=x2**2
```

#### Part 3 - Create the $Y$ and $Z$ coordinates

```python
Y2=g*np.cos(T)
Z2=g*np.sin(T)
```

#### Part 4 - Plot the Wireframe

```python
# Plot the function g(x) in 3-d
fig=plt.figure(figsize=(12,12))
ax=fig.add_subplot(projection='3d') # Use a 3-d projection to plot the surface
ax.plot(x2,g,zs=0,zdir='z',color='k') # plot the function in the (x,y)-plaane i.e. z=0 (See Part 3)

# Plot the wireframe
ax.plot_wireframe(X2,Y2,Z2,linewidth=1,cstride=20,rstride=20,alpha=0.5,color='steelblue') # Plot the surfcae of revolution generated by the curve


# Labels and Viewing angle
ax.set_xlabel('$x$') # Set the x-label 
ax.set_ylabel('$y$') # Set the y-label
ax.set_zlabel('$z$') # Set the z-label

ax.view_init(45,35) # Rotate your view of the solid

plt.show()
```

* In this example the function __plot_wireframe()__ has the follwoing arguments
    1. __X2__ the $x$-ccordinate of a point on the surface
    2. __Y2__ the $y$-coordinatee of a point on the surface
    3. __Z2__ the $z$-ccordinate of a point on the surface
    4. __linewidth__ the width of the  lines on the wireframe
    5. __rstride__ and __cstride__ the separation of the lines on the wireframe
    6. __alpha__ the tranparency
    7. __color__ the color of the wireframe  
    (a full list of Python colors is available at https://matplotlib.org/examples/color/named_colors.html)
